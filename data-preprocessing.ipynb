```
import nltk
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk import word_tokenize
```

```
nltk.download('punkt')
nltk.download('stopwords')
```

```
df=pd.read_csv('names.csv')
```

```
df=df.dropna()
```

```
df
```

```
names=df['name']
```

```
names
```

```
import re
```

```
name=[]
```

```
for i in names:
  j=re.sub('[^a-zA-Z]',' ',i)
  j=j.lower()
  name.append(j)


```

```
name
```

```
s=stopwords.words('english')
```

```
s.append('age')
```

```
s.append('s')
s.append('o')
s.append('dr')
s.append('w')
s.append('son')
s.append('daughter')
s.append('or')
s.append('smt')
s.append('sri')
s.append('md')
```

```
nam=[]
for i in name:
  j=word_tokenize(i)
  words=[k for k in j if k not in s]
  if len(words)>2:
    words=words[0:2]
  nae=''.join(words)
  nam.append(nae)
```

```
labels=list(l)
```
